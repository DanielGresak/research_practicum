```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("weather.csv")
```

```python
df['dt'] = pd.to_datetime(df['dt'],unit='s')
```

```python
df = df.drop(['dt_iso', 'city_name'], axis=1)
```

```python
df['dt'] = pd.to_datetime(df['dt'].astype(str)) + pd.DateOffset(hours=1)
```

```python
df = df.drop(['timezone'], axis = 1)
```
